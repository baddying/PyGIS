一、基于pyecharts的地铁数据可视化

1.引入所需类库

```python
import requests
import json
import pprint
import math
from pyecharts.charts import BMap
from pyecharts import options as opts
from pyecharts.globals import BMapType, ChartType, SymbolType
from IPython.display import IFrame
```

2.获取南宁地铁路线数据

```python
url='https://map.amap.com/service/subway?_1704164391304&srhdata=4501_drw_nanning.json'
response=requests.get(url)
# print(response.text)
result=json.loads(response.text)
# print(result)
stations=[]
for line in result['l']:
    station=[]
    for st in line['st']:
        station.append([float(loc) for loc in st['sl'].split(',')])#将经纬度转换为float类型
    stations.append(station)
print(stations)
```

3.火星坐标转换为百度坐标

```python
pi=3.1415926535897932384626
r_pi=pi*3000.0/180.0
def gcj02_bd09(lng,lat):
    x=lng
    y=lat
    z=math.sqrt(x*x+y*y)+0.00002*math.sin(y*r_pi)
    theta=math.atan2(y,x)+0.000003*math.cos(x*r_pi)
    bd_lng=z*math.cos(theta)+0.0065
    bd_lat=z*math.sin(theta)+0.006
    return bd_lng,bd_lat

bdXY=[]
for line in stations:
    bdXY.append([gcj02_bd09(lng,lat) for lng,lat in line])
print(bdXY)
```

4.利用百度地图api展示地铁路线

```python
map_b=(
    BMap(init_opts=opts.InitOpts(width='800px',height='600px')).add_schema(
        baidu_ak='APVLFO1yNkwRbOmgKn2B0CHW50d5IDMN',
        center=[108.373351,22.823037],
        zoom=12,
        is_roam=True).add(series_name='南宁地铁站线路',
                          type_=ChartType.LINES,
                          data_pair=bdXY,
                          is_polyline=True,
                          linestyle_opts=opts.LineStyleOpts(color='red',width=2,opacity=0.5),
                          ))
map_b.render('南宁地铁分布图.html')
```
