二、航班数据可视化

1.引入所需类库

```python
import numpy as np
import pandas as pd 
import os
from pyecharts import options as opts
from pyecharts.charts import Map,Map3D
from pyecharts.globals import ChartType
```

2.定义三维地图

```python
def map3d_with_bar3d(data) -> Map3D:
    c = (
        Map3D().add_schema(itemstyle_opts=opts.ItemStyleOpts(
                color="rgb(15,101,123)",
                opacity=1,
                border_width=0.8,
                border_color="rgb(62,215,213)",
            ),
            map3d_label=opts.Map3DLabelOpts(is_show=False
            ),
            emphasis_label_opts=opts.LabelOpts(
                is_show=False,
                color="#fff",
                font_size=10,
                background_color="rgba(0,23,11,0)",
            ),
            light_opts=opts.Map3DLightOpts(
                main_color="#fff",
                main_intensity=1.2,
                main_shadow_quality="high",
                is_main_shadow=False,
                main_beta=10,
                ambient_intensity=0.3,
            ),
        )
        .add(
            series_name="",
            data_pair=data,
            type_=ChartType.BAR3D,
            bar_size=1,
            shading="lambert",
            label_opts=opts.LabelOpts(
                is_show=False
            ),
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title="全国航班数量分布图",
                subtitle=""
            ),
            visualmap_opts=opts.VisualMapOpts(is_piecewise=True,
            pieces=[
                {"min": 0, "max": 9, "label": "0-9", "color": "#FFCCCC"},
                {"min": 10, "max": 99, "label": "10-99", "color": "#FF9999"},
                {"min": 100, "max": 199, "label": "100-199", "color": "#FF6666"},
                {"min": 200, "max": 499, "label": "200-499", "color": "#FF3333"},
                {"min": 500, "max": 799, "label": "500-799", "color": "#FF0000"},
                {"min": 800, "label": ">800", "color": "#CC0000"}
            ])))
        
    return c
```

3.读取航班数据

```python
df0=pd.read_excel('./airlines.xls')
size=df0['出发城市'].value_counts()
# print(size)
df1=pd.DataFrame({'航班数量':size})
df2=df0[['出发城市','出发城市y','出发城市x']].drop_duplicates()
df2.index=df2['出发城市']
df2['航班数量']=df1

# print(df2)

city_lines=list(zip(df2['出发城市'],list(zip(df2['出发城市x'],df2['出发城市y'],df2['航班数量']))))

print(city_lines)
```

4.调用函数绘制三维专题图

```python
airport3d=map3d_with_bar3d(city_lines)
airport3d.render(path='航班三维分布图.html')
```
