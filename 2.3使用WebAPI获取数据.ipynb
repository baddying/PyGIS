1.利用高德地图API获取南宁市地铁坐标数据

```python
import requests
import json
url='https://map.amap.com/service/subway?_1704164391304&srhdata=4501_drw_nanning.json' #南宁地铁站点信息网址
response=requests.get(url) #获取网页信息
result=json.loads(response.text) #将网页信息转换为json格式
print(result) #打印json格式的信息
stNames=[]#创建一个空列表用于存放地铁站点信息
stLocs=[]#创建一个空列表用于存放地铁站点坐标信息 
for line in result['l']:#遍历json格式的信息
    for station in line['st']: #遍历json格式的信息
        if station['n'] not in stNames:   #如果地铁站点信息不在列表中
            stNames.append(station['n']) #将地铁站点信息添加到列表中
            stLocs.append(station['sl']) #将地铁站点坐标信息添加到列表中

print(stNames.__len__()) #打印列表长度
print(stNames) #打印站名列表
print(stLocs) #打印坐标列表
```
